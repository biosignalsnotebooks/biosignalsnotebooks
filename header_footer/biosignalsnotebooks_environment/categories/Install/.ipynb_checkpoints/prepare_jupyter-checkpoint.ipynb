<p class="steps">3 - Open a "Command Prompt"</p>
<i>If you are a Microsoft Windows native, just type click on Windows logo (bottom-left corner of the screen) and type "cmd". Then press "Enter".</i>

<p class="steps">4 - Type <span class="color7">"python -m pip install --upgrade pip"</span> (or <span class="color7">"python3 -m pip install --upgrade pip"</span> if you had a previously installed version 2 of Python) to update a very practical <span class="color1">Python</span> tool (<a href="https://pypi.org/project/pip/" target="_blank">pip <img src="../../images/icons/link.png" width="10px" height="10px" style="display:inline"></a>) for installing new packages</p>

<img src="../../images/other/jupyter_install_cmd_final.gif">
